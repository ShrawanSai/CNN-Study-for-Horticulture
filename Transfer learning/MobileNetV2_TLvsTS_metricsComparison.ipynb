```python
import numpy as np
import matplotlib.pyplot as plt
import torch

def plot_metrics(mobilenetv2TL,mobilenetv2TS,mobilenetv2TL_npy,mobilenetv2TS_npy):
    mobilenetv2TL_data = np.load(mobilenetv2TL)
    mobilenetv2TS_data = np.load(mobilenetv2TS)

    # VGG16TL_data = np.load(VGG16TL)
    # MobileNetV2TL_data = np.load(MobileNetV2TL)

    colors = ['blue', 'red']

    epochs_mobilenetv2TL = [i for i in range(1,mobilenetv2TL_npy+1)]
    epochs_mobilenetv2TS = [i for i in range(1,mobilenetv2TS_npy+1)]
    # epochs_VGG16TL = [i for i in range(1,VGG16TL_npy+1)]
    # epochs_MobileNetV2 = [i for i in range(1,MobileNetV2TL_npy+1)]

    mobilenetv2TL_train_loss = mobilenetv2TL_data[:, 1][:mobilenetv2TL_npy]
    mobilenetv2TL_val_loss = mobilenetv2TL_data[:, 2][:mobilenetv2TL_npy]
    mobilenetv2TL_train_accuracy = mobilenetv2TL_data[:, 3][:mobilenetv2TL_npy]
    mobilenetv2TL_val_accuracy = mobilenetv2TL_data[:, 4][:mobilenetv2TL_npy]

    mobilenetv2TS_train_loss = mobilenetv2TS_data[:, 1][:mobilenetv2TS_npy]
    mobilenetv2TS_val_loss = mobilenetv2TS_data[:, 2][:mobilenetv2TS_npy]
    mobilenetv2TS_train_accuracy = mobilenetv2TS_data[:, 3][:mobilenetv2TS_npy]
    mobilenetv2TS_val_accuracy = mobilenetv2TS_data[:, 4][:mobilenetv2TS_npy]

    # VGG16TL_train_loss = VGG16TL_data[:, 1][:VGG16TL_npy]
    # VGG16TL_val_loss = VGG16TL_data[:, 2][:VGG16TL_npy]
    # VGG16_train_accuracy = VGG16TL_data[:, 3][:VGG16TL_npy]
    # VGG16_val_accuracy = VGG16TL_data[:, 4][:VGG16TL_npy]

    # MobileNetV2TL_train_loss = MobileNetV2TL_data[:, 1][:MobileNetV2TL_npy]
    # MobileNetV2TL_val_loss = MobileNetV2TL_data[:, 2][:MobileNetV2TL_npy]
    # MobileNetV2TL_train_accuracy = MobileNetV2TL_data[:, 3][:MobileNetV2TL_npy]
    # MobileNetV2TL_val_accuracy = MobileNetV2TL_data[:, 4][:MobileNetV2TL_npy]

    plt.figure(figsize=(12, 5))
    plt.subplot(121)
    plt.plot(epochs_mobilenetv2TL, mobilenetv2TL_train_loss, label='MobileNetV2 TransferLearning Train Loss',color=colors[0])
    plt.plot(epochs_mobilenetv2TL, mobilenetv2TL_val_loss, linestyle='--', label='MobileNetV2 TransferLearning Val Loss',color=colors[0])
    plt.plot(epochs_mobilenetv2TS, mobilenetv2TS_train_loss, label='MobileNetV2 Train_Scratch Train Loss',color=colors[1])
    plt.plot(epochs_mobilenetv2TS, mobilenetv2TS_val_loss, linestyle='--', label='MobileNetV2 Train_Scratch Val Loss',color=colors[1])
    # plt.plot(epochs_VGG16TL, VGG16TL_train_loss, label='VGG16 TL Train Loss',color=colors[1])
    # plt.plot(epochs_VGG16TL, VGG16TL_val_loss, linestyle='--', label='VGG16 TL Val Loss',color=colors[1])
    # plt.plot(epochs_MobileNetV2, MobileNetV2TL_train_loss, label='MobileNetV2 TL Train Loss',color=colors[2])
    # plt.plot(epochs_MobileNetV2, MobileNetV2TL_val_loss, linestyle='--', label='MobileNeV2 TL Val Loss',color=colors[2])
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.title('Train and Validation Loss for TransferLearning Model vs Train-Scratch model(Dataset3)',fontsize=8)
    plt.legend()

    plt.subplot(122)
    plt.plot(epochs_mobilenetv2TL, mobilenetv2TL_train_accuracy, label='MobileNetV2 TransferLearning Train Accuracy',color=colors[0])
    plt.plot(epochs_mobilenetv2TL, mobilenetv2TL_val_accuracy, linestyle='--', label='MobileNetV2 TransferLearning Val Accuracy',color=colors[0])
    plt.plot(epochs_mobilenetv2TS, mobilenetv2TS_train_accuracy, label='MobileNetV2 Train_Scratch Train Accuracy',color=colors[1])
    plt.plot(epochs_mobilenetv2TS, mobilenetv2TS_val_accuracy, linestyle='--', label='MobileNetV2 Train_Scratch Val Accuracy',color=colors[1])
    # plt.plot(epochs_VGG16TL, VGG16_train_accuracy, label='VGG16 TL Train Accuracy',color=colors[1])
    # plt.plot(epochs_VGG16TL, VGG16_val_accuracy, linestyle='--', label='VGG16 TL Val Accuracy',color=colors[1])
    # plt.plot(epochs_MobileNetV2, MobileNetV2TL_train_accuracy, label='MobileNetV2 TL Train Accuracy',color=colors[2])
    # plt.plot(epochs_MobileNetV2, MobileNetV2TL_val_accuracy, linestyle='--', label='MobileNetV2 TL Val Accuracy',color=colors[2])
    plt.xlabel('Epoch')
    plt.ylabel('Accuracy')
    plt.title('Train and Validation Accuracy for TransferLearning Model vs Train-Scratch Model(Dataset3)',fontsize=8)
    plt.legend()

    plt.show()
```
